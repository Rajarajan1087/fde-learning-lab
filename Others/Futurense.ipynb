```python
def sum(a,b):
  return a+b

def subtract_numbers(a,b):
  if a<b:
    return b-a
  else:
    return a-b

def multiply_numbers(a,b):
  return a*b

def divide_numbers(a,b):
  if(b==0):
    print("Enter Divisor as non zero number")
  else:
    return a/b

s = True

while s:

  choice = input("""Select operation:
1. Add
2. Subtract
3. Multiply
4. Divide
5. Exit

Enter choice (1/2/3/4/5):
""")
  if choice == '5':
    break
  else:
    try:
      a = int(input("Enter first number :"))
      b = int(input("Enter second number :"))
    except ValueError:
      print("enter whole number only")
      continue

    if choice == '1':
      print("Addition:",sum(a,b))
    elif choice == '2':
      print("subtraction:",subtract_numbers(a,b))
    elif choice == '3':
      print("Multiply:",multiply_numbers(a,b))
    elif choice == '4':
      print("Division:",divide_numbers(a,b))
```
