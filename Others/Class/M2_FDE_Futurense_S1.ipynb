```python
# ETL Processing

import pandas as pd

# -----------------------------------
# EXTRACT STEP
# Raw Swiggy Order Data
# -----------------------------------

raw_data = {
    "Order_ID": [101, 102, 103, 104],
    "City": ["Delhi", "delhi", "DELHI", "Mumbai"],
    "Amount": [450, 500, None, 700],
    "Delivery_Time": [32, 29, 40, -5]
}

df = pd.DataFrame(raw_data)

print("\nRAW DATA")
print(df)

# -----------------------------------
# TRANSFORM STEP
# Cleaning the data
# -----------------------------------

# Standardize city names
df["City"] = df["City"].str.title()

# Replace missing amount with 0
df["Amount"] = df["Amount"].fillna(0)

# Remove invalid delivery time
df = df[df["Delivery_Time"] > 0]

print("\nCLEANED DATA")
print(df)

# -----------------------------------
# LOAD STEP
# Save cleaned data
# -----------------------------------

df.to_csv("cleaned_swiggy_orders.csv", index=False)

print("\nData Loaded into Warehouse (CSV File)")
```
