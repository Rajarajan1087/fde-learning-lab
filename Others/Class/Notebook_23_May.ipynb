```python
import pandas as pd

# -----------------------------------
# EXTRACT STEP
# Raw Swiggy Order Data
# -----------------------------------

raw_data = {
    "Order_ID": [101, 102, 103, 104],
    "City": ["Delhi", "delhi", "DELHI", "Mumbai"],
    "Amount": [450, 500, None, 700],
    "Delivery_Time": [32, 29, 40, -5]
}

df = pd.DataFrame(raw_data)

print("\nRAW DATA")
print(df)

# -----------------------------------
# TRANSFORM STEP
# Cleaning the data
# -----------------------------------

# Standardize city names
df["City"] = df["City"].str.title()

# Replace missing amount with 0
df["Amount"] = df["Amount"].fillna(0)

# Remove invalid delivery time
df = df[df["Delivery_Time"] > 0]

print("\nCLEANED DATA")
print(df)

# -----------------------------------
# LOAD STEP
# Save cleaned data
# -----------------------------------

df.to_csv("cleaned_swiggy_orders.csv", index=False)

print("\nData Loaded into Warehouse (CSV File)")
```

```python
import pandas as pd

# -----------------------------------
# EXTRACT + LOAD
# Store raw Netflix data first
# -----------------------------------

netflix_data = {
    "User": ["A", "B", "C"],
    "Movie": ["Interstellar", "Friends", "Narcos"],
    "Pause_Count": [5, 0, 12],
    "Watch_Percentage": [98, 100, 35],
    "Search_History": ["Sci-Fi", "Comedy", "Crime"]
}

raw_df = pd.DataFrame(netflix_data)

print("\nRAW DATA STORED")
print(raw_df)

# Save raw data directly
raw_df.to_csv("raw_netflix_data.csv", index=False)

# -----------------------------------
# TRANSFORM LATER
# AI analysis after storage
# -----------------------------------

raw_df["Likely_To_Leave"] = raw_df["Watch_Percentage"] < 50

print("\nAFTER AI ANALYSIS")
print(raw_df)
```

```python
import pandas as pd

# -----------------------------------
# OLD DATABASE
# -----------------------------------

old_messages = {
    "Message_ID": [1, 2, 3],
    "Message": ["Hi", "How are you?", "Meeting at 5"]
}

old_df = pd.DataFrame(old_messages)

print("\nOLD DATA")
print(old_df)

# -----------------------------------
# NEW INCOMING DATA
# -----------------------------------

new_messages = {
    "Message_ID": [4, 5],
    "Message": ["Project done", "See you tomorrow"]
}

new_df = pd.DataFrame(new_messages)

print("\nNEW DATA")
print(new_df)

# -----------------------------------
# INCREMENTAL LOAD
# Only new records added
# -----------------------------------

updated_df = pd.concat([old_df, new_df])

print("\nUPDATED DATABASE")
print(updated_df)
```

```python
import pandas as pd

# -----------------------------------
# REAL-TIME RIDE REQUESTS
# -----------------------------------

uber_data = {
    "Location": ["Airport", "Airport", "City", "Airport"],
    "Drivers_Available": [2, 1, 10, 1],
    "Ride_Requests": [20, 25, 8, 30]
}

df = pd.DataFrame(uber_data)

print("\nUBER LIVE DATA")
print(df)

# -----------------------------------
# AI SURGE LOGIC
# -----------------------------------

df["Surge_Multiplier"] = (
    df["Ride_Requests"] / df["Drivers_Available"]
)

print("\nSURGE PRICING")
print(df)
```
