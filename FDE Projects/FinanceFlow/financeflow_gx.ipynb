## Cell 5 - Register dataset + build pre-built Expectation Suite

These are the 6 rules already written for the Financial Flow SMB data.  

1.	credit_score must not be null.
2.	credit_score must be between 300 and 850 (valid FICO range).
3.	default_flag must not be null.
4.	annual_revenue_usd must be greater than or equal to 0.
5.	debt_to_income_ratio must be between 0.05 and 1.00.
6.	sector must be one of: Retail, Manufacturing, Technology, Hospitality, Healthcare, Construction, Agriculture.

```python
datasource_name = "FinFlow_datasource"

# Check if the datasource already exists, if so, retrieve it; otherwise, add it.
if datasource_name in context.data_sources.all():
    data_source = context.data_sources.get(datasource_name)
else:
    data_source = context.data_sources.add_pandas(datasource_name)

data_asset_name = "FinFlow_Loan"
# Check if the data asset already exists, if so, retrieve it; otherwise, add it.
if data_asset_name in data_source.get_asset_names():
    data_asset = data_source.get_asset(data_asset_name)
else:
    data_asset = data_source.add_dataframe_asset(name=data_asset_name)

batch_definition_name = "FinFlow_full_batch"
# Attempt to add the batch definition; if it already exists, retrieve it.
try:
    batch_definition = data_asset.add_batch_definition_whole_dataframe(batch_definition_name)
except ValueError as e:
    if f'"{batch_definition_name}" already exists' in str(e):
        print(f"Batch definition '{batch_definition_name}' already exists. Retrieving it.")
        # Iterate through the list of batch definitions to find the one with the correct name
        batch_definition = None
        for bd in data_asset.batch_definitions:
            if bd.name == batch_definition_name:
                batch_definition = bd
                break
        if batch_definition is None:
            raise RuntimeError(f"Could not find existing batch definition '{batch_definition_name}'")
    else:
        raise # Re-raise if it's a different ValueError

# --- Pre-built Expectation Suite ---
suite_name = "FinFlow_suite"
# Attempt to delete the suite first if it exists, to ensure a clean slate
try:
    context.suites.delete(suite_name)
    print(f"Existing suite '{suite_name}' deleted for a fresh definition.")
except Exception as e:
    # If the suite doesn't exist, an error will be raised, which is fine.
    # We catch a general exception to be robust to different error messages from GX.
    if f"Cannot delete ExpectationSuite with name {suite_name} because it does not exist." in str(e) or f"The ExpectationSuite '{suite_name}' does not exist." in str(e):
        print(f"Suite '{suite_name}' does not exist, proceeding to create.")
    else:
        # Re-raise if it's a different error that we don't expect
        raise

# Now, add the suite, guaranteed not to exist from prior operations (or handled by the try-except)
suite = context.suites.add(gx.ExpectationSuite(name=suite_name))
print(f"Created new suite: {suite_name}")


## Expectations code below:

# credit_score: expect not null
# This is the target variable. A churn model cannot be trained
# if 31% of the labels are missing.
suite.add_expectation(
    gx.expectations.ExpectColumnValuesToNotBeNull(column="credit_score")
)

# credit_score: expect values between 300 and 850
suite.add_expectation(
    gx.expectations.ExpectColumnValuesToBeBetween(
        column="credit_score",
        min_value=300,
        max_value=850
    )
)



# default_flag : expect not null
# This is the target variable. A churn model cannot be trained
# if 31% of the labels are missing.
suite.add_expectation(
    gx.expectations.ExpectColumnValuesToNotBeNull(column="default_flag")
)

# annual_revenue_usd: expect values >= 0
suite.add_expectation(
    gx.expectations.ExpectColumnValuesToBeBetween(
        column="annual_revenue_usd",
        min_value=0
    )
)

#debt_to_income_ratio must be between 0.05 and 1.00.
suite.add_expectation(
    gx.expectations.ExpectColumnValuesToBeBetween(
        column="debt_to_income_ratio",
        min_value=0.05,
        max_value=1.00
    )
)

# sector must be one of: Retail, Manufacturing, Technology, Hospitality, Healthcare, Construction, Agriculture.
suite.add_expectation(
    gx.expectations.ExpectColumnValuesToBeInSet(
        column="sector",
        value_set=["Retail", "Manufacturing", "Technology", "Hospitality", "Healthcare", "Construction", "Agriculture"]
    )
)

print("✅ Pre-built checkpoint ready — 6 expectations loaded")
```

## Cell 6 — Run the checkpoint

> **Trainer:** run this cell live. The results will show which rules pass and which fail.

```python
validation_definition = context.validation_definitions.add(
    gx.ValidationDefinition(
        name="FinFlow_validation",
        data=batch_definition,
        suite=suite,
    )
)

checkpoint = context.checkpoints.add(
    gx.Checkpoint(
        name="FinFlow_checkpoint",
        validation_definitions=[validation_definition],
    )
)

results = checkpoint.run(batch_parameters={"dataframe": df})

print(results)
```
