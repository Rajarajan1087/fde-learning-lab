•	**How many rows are missing default_flag? Why can these rows not be used in model training?**



1.   25 values missing out of 400 ( 6.2% missing )
2.   These rows cannot be used because default_flag is the target variable (label) for supervised learning.

      In supervised classification, the model learns by mapping input features → known outcomes (default vs. no-default)

      Without default_flag, the model has no label to learn from — it cannot determine what patterns indicate default

      These rows would be excluded from training (typically dropped) since they provide no supervision signal

      Using them would introduce label noise and degrade model performance
