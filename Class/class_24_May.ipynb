```python
df = pd.DataFrame({'city':['delhi', 'DELHI','Mumbai', 'MUMBAI','DELhi'], 'Order_ID':[101,102,304,111,222]})
df
```

```python
df_new = df.copy() # copy function will ensure it creases a seperate copy at different memory address
df_new
```

```python
df_new['city'] = df_new['city'].str.upper()
df_new
```

```python
df_new['city'] = df_new['city'].map({'DELHI':0,'MUMBAI':1})
df_new
```
